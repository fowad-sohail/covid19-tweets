```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
CSV_PATH = '/content/drive/My Drive/Colab Notebooks/covid19_tweets.csv'
df = pd.read_csv(CSV_PATH)
```

```
df.info()
```

```
df = df.dropna() # Drop N/A's
```

```
# Inspect which months the csv includes
pd.DatetimeIndex(df['date']).month.unique()
```

```
num_tweets_by_location = df['user_location'].value_counts()
```

```
print(num_tweets_by_location)
```

```
fig, ax = plt.subplots(figsize = (13,5))
plt.xlabel("Location", fontsize = 12)
plt.ylabel("NO. Tweets", fontsize = 12)
num_tweets_by_location[0:50].plot(kind='bar', title = "Top 15 Countries Posting about Covid-19 in July and August", )
```
